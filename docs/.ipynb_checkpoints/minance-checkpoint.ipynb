```python
class Account(object):
    """
    This is an account class containing user information and balances
    """
    user_name = None
    balance = None
    
    def __init__(self, user_name):
        self.logger = logging.getLogger("root")
        self.user_name = user_name
        self.balance = {}
    
    def deposit(self, amount, ticker):
        if self.balance.get(ticker):
            self.balance[ticker] += amount
        else:
            self.balance.update({ticker: amount})
            
    def withdraw(self, amount, ticker):
        if self.balance.get(ticker) >= amount:
            self.balance[ticker] -= amount
        else:
            self.logger.error("The {} balance is not sufficient. Withdraw unsuccessful".format(ticker))
            raise Exception
            
    def buy(self, qty, share):
        buy(self, qty, share)
        
    def sell(self, qty, share):
        sell(self, qty, share)
```
